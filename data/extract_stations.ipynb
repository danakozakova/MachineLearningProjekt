```python
import pandas as pd

path = "stations.txt"

with open(path, encoding="utf-8") as f:
    lines = f.readlines()

# riadky 0–3 sú hlavička, od 4 ďalej sú dáta (overené z tvojho výstupu)
data_lines = lines[4:]

rows = []
for ln in data_lines:
    if not ln.strip():
        continue

    site      = ln[0:7].strip()
    site_name = ln[13:53].strip()
    start     = ln[58:62].strip()
    end       = ln[66:70].strip()
    lat_str   = ln[71:79].strip()
    lon_str   = ln[80:89].strip()

    rows.append({
        "Site": site,
        "Site_name": site_name,
        "Start": start,
        "End": end,
        "Lat_str": lat_str,
        "Lon_str": lon_str,
    })

df_tmp = pd.DataFrame(rows)

# prekonvertuj na čísla a nechaj len to, čo potrebuješ
df = df_tmp.copy()
df["Lat"] = pd.to_numeric(df["Lat_str"], errors="coerce")
df["Lon"] = pd.to_numeric(df["Lon_str"], errors="coerce")

df = df[["Site", "Site_name", "Lat", "Lon"]]
df = df.dropna(subset=["Lat", "Lon"])

df.head(), df.shape
```

```python
df.to_csv("bom_stations.csv", index=False)
```
